```
import requests
import csv
import pandas as pd
import urllib.request 
import pylab
import numpy as math
import io
import matplotlib.pyplot as plt
import functools 


#This is a list of all stations in Madagascar. 
Madagascar  = ['67009_195109_201712.txt','67012_199901_201712.txt','67019_196104_199412.txt','67023_200901_201712.txt','67025_196010_201708.txt','67027_195109_201712.txt','67037_200901_201311.txt','67067_200801_200912.txt','67072_200801_200905.txt','67073_195109_201708.txt','67083_197501_201712.txt','67085_195109_201612.txt','67095_195109_201712.txt','67107_200901_201310.txt','67113_200901_200912.txt','67117_199901_200806.txt','67131_200801_200912.txt','67137_197111_201712.txt','67143_196201_201712.txt','67152_200102_201708.txt','67157_200906_201708.txt','67161_195109_201708.txt','67197_195109_201612.txt']

Mad_df = pd.DataFrame()

for i in range(len(Madagascar)):
  url_str = ('https://opendata.dwd.de/climate_environment/CDC/observations_global/CLIMAT/monthly/qc/precipitation_total/historical/' + Madagascar[i])
  urllib.request.urlretrieve(url_str,"data.txt")
  df = pd.read_csv("data.txt",delimiter=';')
  df.drop(df.loc[df['Jahr']<1980].index, inplace=True)
  df['average'] = df.mean(numeric_only=True, axis=1)
  average = df[['Jahr','average']].copy()
  Mad_df = Mad_df.append(average, ignore_index=True)


#Mad_df.to_csv('Mad.csv') # to be deleted 
#depending on whether this is needed for further visualization. Was established for understanding how data looks like
Mad_df.head
avg_rainfall = Mad_df.groupby(['Jahr']).mean()
avg_rainfall.plot(c = 'blue', label = 'Average rainfall')

```
